# The following will download the farmers market file, create a Pandas dataframe, select the cities with the coordinates, and map them using Kepler

```python
%%bash

if [ -e Export.csv ]
then
    echo "File exists, skipping for now"
else
    wget https://raw.githubusercontent.com/Msanghi/NCData4Good/master/Export.csv
fi




```

```python
import pandas as pd
pd.set_option('display.max_rows', 500)
pd.set_option('display.max_columns', 500)
pd.set_option('display.width', 1000)
from keplergl import KeplerGl
map_1 = KeplerGl(height=1000)
```

```python
farmDF = pd.read_csv("Export.csv", delimiter = ',', encoding='latin-1')
```

```python
display(farmDF)
```

```python
farmlocationDF = farmDF[['MarketName', 'x', 'y']]
```

```python
mapfarmDF = farmlocationDF.rename(index=str, columns={"y": "Latitude", "x": "Longitude"})
```

```python
mapfarmDF
```

```python
from keplergl import KeplerGl
map_1 = KeplerGl(height=1024, length=768)
```

```python
map_1.add_data(data=mapfarmDF, name='Farmers Markets')
```
